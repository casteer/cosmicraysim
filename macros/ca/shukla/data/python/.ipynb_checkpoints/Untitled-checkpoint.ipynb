# Comparison of QGSP_BERT_HP and Shielding Physics Lists

```python
%matplotlib inline

import matplotlib
import numpy as np
import matplotlib.pyplot as plt

filenames=['../shaft_air_QGSP_BERT_HP.dat', '../shaft_air_shielding_rates.dat']

def plot_data(filename,fomt='ro'):
    positions = []
    rate = []
    err_rate = []
    for line in open(filename):
        positions.append(float(line.split()[0])/100.0)
        rate.append(float(line.split()[1])*60.0*60.0/1000.0)
        err_rate.append(float(line.split()[2])*60.0*60.0/1000.0)
    plt.errorbar(positions,rate,yerr=err_rate,fmt=fomt)
    plt.grid(True)
    plt.xlabel('Position (m)')
    plt.ylabel('Rate (x1000, hour$^{-1}$)')

plt.figure(num=None, figsize=(8, 6), dpi=80, facecolor='w', edgecolor='k')
plot_data(filenames[0],fomt='ro')
plot_data(filenames[1],fomt='go')
```

# Study of background rock density and rates

```python
%matplotlib inline

import matplotlib
import numpy as np
import matplotlib.pyplot as plt

filenames=[
    '../shaft_air_shielding_density2p0_rates.dat', 
    '../shaft_air_shielding_density2p2_rates.dat', 
    '../shaft_air_shielding_density2p4_rates.dat', 
    '../shaft_air_shielding_rates.dat']

def plot_data(filename,fomt='ro'):
    positions = []
    rate = []
    err_rate = []
    for line in open(filename):
        positions.append(float(line.split()[0])/100.0)
        rate.append(float(line.split()[1])*60.0*60.0/1000.0)
        err_rate.append(float(line.split()[2])*60.0*60.0/1000.0)
    plt.errorbar(positions,rate,yerr=err_rate,fmt=fomt)
    plt.grid(True)
    plt.xlabel('Position (m)')
    plt.ylabel('Rate (x1000, hour$^{-1}$)')

plt.figure(num=None, figsize=(8, 6), dpi=80, facecolor='w', edgecolor='k')
plot_data(filenames[0],fomt='ro')
plot_data(filenames[1],fomt='go')
plot_data(filenames[2],fomt='bo')
plot_data(filenames[3],fomt='mo')
```

# Air vs Water Shaft Fill Comparison

```python
%matplotlib inline

import matplotlib
import numpy as np
import matplotlib.pyplot as plt

filenames=[
    '../shaft_air_rates_50k_rates.dat', 
    '../shaft_air_fill4m_rates_50k_rates.dat', 
    '../shaft_air_fill8m_rates_50k_rates.dat',
    '../shaft_water_rates_50k_rates.dat']

def plot_data(filename,fomt='ro'):
    positions = []
    rate = []
    err_rate = []
    for line in open(filename):
        positions.append(float(line.split()[0])/100.0)
        rate.append(float(line.split()[1])*60.0*60.0*24.0/1000.0)
        err_rate.append(float(line.split()[2])*60.0*60.0*24.0/1000.0)
    plt.errorbar(positions,rate,yerr=err_rate,fmt=fomt)
    plt.grid(True)
    plt.xlabel('Position (m)')
    plt.ylabel('Rate (x1000, hour$^{-1}$)')

plt.figure(num=None, figsize=(8, 6), dpi=80, facecolor='w', edgecolor='k')
plot_data(filenames[0],fomt='ro')
plot_data(filenames[1],fomt='go')
plot_data(filenames[2],fomt='bo')
```
